```python
from lms import LMS, LmsType
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Exercise 1 - Alps Water Dataset

```python
X = [
[194.5],
[194.3],
[197.9],
[198.4],
[199.4],
[199.9],
[200.9],
[201.1],
[201.4],
[201.3],
[203.6],
[204.6],
[209.5],
[208.6],
[210.7],
[211.9],
[212.2]
]

Y = [
[20.79],
[20.79],
[22.40],
[22.67],
[23.15],
[23.35],
[23.89],
[23.99],
[24.02],
[24.01],
[25.14],
[26.57],
[28.49],
[27.76],
[29.04],
[29.88],
[30.06]
]
```

```python
lms = LMS()

lms.fit(x=X, y=Y, type=LmsType.Linear)
y_predict_linear = lms.predict(x=X)
y_200_linear = lms.predict(x=[[200]])

lms.fit(x=X, y=Y, type=LmsType.Quadratic)
y_predict_quadratic = lms.predict(x=X)
y_200_quadratic = lms.predict(x=[[200]])
```

```python
plt.scatter(X, Y, color="black", label="Data")
plt.plot(X, y_predict_linear, color="brown", label="Linear Predict")
plt.plot(X, y_predict_quadratic, color="orange", label="Quadratic Predict")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('LMS')
plt.legend()
plt.grid()
```

```python
print("BPt(200) ->\n   LMS Linear:    %s\n   Lms Quadratic: %s" %(y_200_linear[0][0], y_200_quadratic[0][0]))
```
